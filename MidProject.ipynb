**Clean the data**

```python
# Remove duplicates
data = data.drop_duplicates()
data
```

```python
# Handle missing values
data = data.dropna(subset=['price', 'year', 'odometer', 'manufacturer'])  # Drop rows with critical nulls
data['condition'].fillna('unknown', inplace=True)  # Fill missing 'condition' with 'unknown
data
```

```python
# Correct data types
data['year'] = pd.to_numeric(data['year'], errors='coerce')
data['price'] = pd.to_numeric(data['price'], errors='coerce')
data
```

```python
# Remove outliers for price
data = data[(data['price'] > 500) & (data['price'] < 100000)]  # Assume reasonable range
data
```

```python
# Remove outliers for odometer
data = data[data['odometer'] < 300000]  # Remove unrealistic mileage
data
```

**Define** **and** **Anser** **Business** **Questions**

1- What are the most popular car brands listed on Craigslist?

```python
import matplotlib.pyplot as plt
popular_brands = data['manufacturer'].value_counts().head(10)
popular_brands.plot(kind='bar', color='skyblue', figsize=(10, 5))
plt.title("Top 10 Most Popular Car Brands")
plt.xlabel("Manufacturer")
plt.ylabel("Number of Listings")
```

2- What is the average price of cars based on their brand?

```python
avg_price_brand = data.groupby('manufacturer')['price'].mean().sort_values(ascending=False).head(10)
avg_price_brand.plot(kind='bar', color='orange', figsize=(10, 5))
plt.title("Average Price by Car Brand")
plt.xlabel("Manufacturer")
plt.ylabel("Average Price ($)")
plt.show()
```

3- How do car prices vary by condition?

```python
import seaborn as sns
sns.boxplot(x='condition', y='price', data=data, palette='coolwarm')
plt.title("Car Prices by Condition")
plt.xlabel("Condition")
plt.ylabel("Price ($)")
plt.show()
```

4- What is the distribution of car listingsby state?

```python
top_states = data['state'].value_counts().head(10)
top_states.plot(kind='bar', color='green', figsize=(10, 5))
plt.title("Top 10 States by Number of Listings")
plt.xlabel("State")
plt.ylabel("Number of Listings")
plt.show()
```

5- What is the relationship between odometer readings and price?

```python
sns.scatterplot(x='odometer', y='price', data=data, alpha=0.5)
plt.title("Price vs Odometer")
plt.xlabel("Odometer (miles)")
plt.ylabel("Price ($)")
plt.show()
```

6- How does car type influence price?

```python
sns.boxplot(x='type', y='price', data=data, palette='muted')
plt.title("Price by Car Type")
plt.xlabel("Type")
plt.ylabel("Price ($)")
plt.xticks(rotation=45)
plt.show()
```

7- What is the most common fuel type?

```python
fuel_type = data['fuel'].value_counts()
fuel_type.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), colors=sns.color_palette("pastel"))
plt.title("Distribution of Fuel Types")
plt.ylabel("")  # Remove y-label for cleaner pie chart
plt.show()
```
